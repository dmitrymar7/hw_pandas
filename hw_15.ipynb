```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})

fields = set(data['whoAmI'])

res = pd.DataFrame(columns=fields)

for i in data['whoAmI']:
    dct = {field: 1 if field == i else 0 for field in fields}
    another_df = pd.DataFrame([dct])
    res = pd.concat([res, another_df], ignore_index=True)
res
```
